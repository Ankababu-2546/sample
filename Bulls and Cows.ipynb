Randomly generate a 4-digit number. 

Ask the user to guess a 4-digit number. 

For every digit that the user guessed correctly in the correct place, they have a “cow”. 

For every digit the user guessed correctly in the wrong place is a “bull.” 

Every time the user makes a guess, tell them how many “cows” and “bulls” they have. 

Once the user guesses the correct number, the game is over. 

Keep track of the number of guesses the user makes throughout teh game and tell the user at the end.

```python
import random
 
secret_number = str(random.randint(10, 99))
print(secret_number)
print("Guess the number. It contains 2 digits.")
 
no_of_tries = 3
 
while no_of_tries > 0:
   player_guess = input("Enter your guess: ")
   
   if player_guess == secret_number:
       print("You guessed it!")
       print("YOU WIN.")
       break
   else:
       bulls = 0
       cows = 0
      
       if player_guess[0] == secret_number[0]:
           bulls += 1
       if player_guess[1] == secret_number[1]:
           bulls += 1
       if player_guess[0] == secret_number[1]:
           cows += 1
       if player_guess[1] == secret_number[0]:
           cows += 1
 
       print("Bulls: ",bulls)
       print("Cows: ",cows)
 
       no_of_tries -= 1
 
       if no_of_tries < 1:
           print("You lost the game.")
           break
```

```python
import random

def generate_secret_number():
    """Generate a random 4-digit secret number."""
    return str(random.randint(1000, 9999))

def get_cows_bulls(secret_number, guess):
    """Calculate the number of cows and bulls for a given guess."""
    cows = 0
    bulls = 0
    for i in range(4):
        if guess[i] == secret_number[i]:
            bulls += 1
        elif guess[i] in secret_number:
            cows += 1
    return cows, bulls

def main():
    print("Welcome to the Cows and Bulls game!")
    secret_number = generate_secret_number()
    print(secret_number)
    attempts = 0
    
    while True:
        guess = input("Enter your guess (a 4-digit number): ")
        
        if len(guess) != 4 or not guess.isdigit():
            print("Please enter a valid 4-digit number.")
            continue
        
        attempts += 1
        cows, bulls = get_cows_bulls(secret_number, guess)
        
        print(f"Cows: {cows}, Bulls: {bulls}")
        
        if bulls == 4:
            print(f"Congratulations! You guessed the number {secret_number} in {attempts} attempts.")
            break

if __name__ == "__main__":
    main()
```

```python
import random

def compare_numbers(number, user_guess):
    cowbull = [0,0] #cows, then bulls
    for i in range(len(number)):
        if number[i] == user_guess[i]:
            cowbull[1]+=1
        else:
            cowbull[0]+=1
    return cowbull

if __name__=="__main__":
    playing = True #gotta play the game
    number = str(random.randint(0,9999)) #random 4 digit number
    guesses = 0

   
    while playing:
        user_guess = input("Give me your best guess!")
        if user_guess == "exit":
            break
        cowbullcount = compare_numbers(number,user_guess)
        guesses+=1

        print("You have "+ str(cowbullcount[0]) + " cows, and " + str(cowbullcount[1]) + " bulls.")

        if cowbullcount[1]==4:
            playing = False
            print("You win the game after " + str(guesses) + "! The number was "+str(number))
            break #redundant exit
        else:
            print("Your guess isn't quite right, try again.")
```
